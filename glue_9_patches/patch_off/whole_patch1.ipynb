```python
def read_off(path):

    with open(path, 'r') as f:
        a = f.read().split()
        b = a[4:] # remove 'COFF' and other lines
        
        # list to array
        file_float = []
        for num in b:
            file_float.append(float(num))
        c = np.array(file_float)
        d = c.reshape(-1,7)
        
        # only keep rgb lines, remove the index of vertex lines
        ind = list(d[:,0]).index(3)
        vertex = d[0:ind,0:3]
        U = d[0:ind,3:6]            
    return vertex,U
```

```python
def b(RGB1,RGB2,RGB3,RGB4,RGB5,RGB6,RGB7,RGB8,RGB9,v1_ind,v2_ind,v3_ind,v4_ind,v5_ind,v6_ind,v7_ind,v8_ind,v9_ind,p1,p2,p3,p4,
      p5,p6,p7,p8,p9,p):
    
    X_u1 = np.zeros((p,p))
    for i in range(0,p1):
        ind = v1_ind[i]
        X_u1[ind][ind] = RGB1[i]
        
    X_u2 = np.zeros((p,p))
    for i in range(0,p2):
        indd = v2_ind[i]
        X_u2[indd][indd] = RGB2[i]

    X_u3 = np.zeros((p,p))
    for i in range(0,p3):
        inddd = v3_ind[i]
        X_u3[indd][indd] = RGB3[i]
        
    X_u4 = np.zeros((p,p))
    for i in range(0,p4):
        indddd = v4_ind[i]
        X_u4[indd][indd] = RGB4[i]    
        
    X_u5 = np.zeros((p,p))
    for i in range(0,p5):
        indddd = v5_ind[i]
        X_u5[indd][indd] = RGB5[i] 
        
    X_u6 = np.zeros((p,p))
    for i in range(0,p6):
        indddd = v6_ind[i]
        X_u6[indd][indd] = RGB6[i] 
        
    X_u7 = np.zeros((p,p))
    for i in range(0,p7):
        indddd = v7_ind[i]
        X_u7[indd][indd] = RGB7[i]  
        
    X_u8 = np.zeros((p,p))
    for i in range(0,p8):
        indddd = v8_ind[i]
        X_u8[indd][indd] = RGB8[i]  
        
    X_u9 = np.zeros((p,p))
    for i in range(0,p9):
        indddd = v9_ind[i]
        X_u9[indd][indd] = RGB9[i]  
    
    Xu = X_u1+X_u2+X_u3+X_u4+X_u5+X_u6+X_u7+X_u8+X_u9
    b_uv = np.zeros(p)
    
    for i in range(0,p):
        b_uv[i] = Xu[i][i]
    
    return b_uv
```
